# 4. Cleaning the Dataset

```python
# Performing these steps ensures that the text is as relevant as possible to the message that the actual 
# review is trying to represent. For example capital I isn’t evaluated as being in the stop word list whereas 
# the lowercase representation is.

# 1. Lowercasing all words
table['GuestLowercase'] = table['GuestReview'].apply(lambda x: " ".join(x.lower() for x in x.split()))
table['HostLowercase'] = table['HostOverview'].apply(lambda x: " ".join(x.lower() for x in x.split()))
# 2. Tokenization = Removing punctuation
table['GuestNopunc'] = table['GuestLowercase'].str.replace('[^\w\s]', '')
table['HostNopunc'] = table['HostLowercase'].str.replace('[^\w\s]', '')
# 3. Removing stopwords #Stop words are commonly occuring words, that hold little to no meaning.
#stopwords = stopwords.words('english')
table['GuestNostop'] = table['GuestNopunc'].apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))
table['HostNostop'] = table['HostNopunc'].apply(lambda x: " ".join(x for x in x.split() if x not in stopwords))

table.head(1)
```

```python
# Lemmatise the Reviews - process of transforming your natural text from english to lemming language. This cuts out 
# the number of words that are available for analysis by combining similar forms, into one base form. One of other 
#processes that is commonly used to cut down the the number of unique words in natural text processing is a process 
# called stemming.Import textblob to lemmatize. Lemmatize final review format

#GUEST - Final Cleaned review
table['GuestCleaned'] = table['GuestNoStop2']\
.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))

#HOST  - Final Cleaned review
table['HostCleaned'] = table['HostNoStop2']\
.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
```
